Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like ``leaflets three, let it be'' for Poisonous Oak and Ivy.***

### Lets check count-plot for all the features

````python
columns = df.columns
````

````python
plt.suptitle('count plot for all features', 
             fontsize=20, 
             fontweight='bold', 
             alpha=0.8, 
             y=1.)

for i in range(0, len(columns)):
    ax = sns.countplot(x=df[columns[i]],palette="Set2") # Palette is color type
    plt.xlabel(columns[i], fontsize=16)
    plt.xticks(rotation=45, fontsize=16)
    ax.set_ylabel('Count', fontsize=16)
    plt.yticks(fontsize=16)
    plt.show()
````

### Let's check the crosstab for all

````python
for i in df.columns:
    if i != "class":
        crosstab_table = pd.crosstab(df[i], df['class'])
        print(crosstab_table)
        print("-"*50)
````

### Let's check chi-square test to test for independence between two categorical variables

````python
import pandas as pd
from scipy.stats import chi2_contingency

for i in df.columns:
    contingency_table = pd.crosstab(df[i], df['class'])
    stat, p, dof, expected = chi2_contingency(contingency_table)
    print(f'Chi-square test: {i}')
    print('p-value:', p)
    print('-'*50)
````

***For all the variables listed, the p-value is less than 0.05. we usually reject the null hypothesis and conclude that there is a statistically significant relationship between the two categorical variables***
